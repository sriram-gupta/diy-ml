```python
import torch
torch.manual_seed(42) # for reproducibility

from torch import nn, sigmoid
from utils import generateXGrid

model_1 = nn.Sequential( nn.Linear(2,1) )
model_2 = nn.Sequential( nn.Linear(2,1), nn.Sigmoid() )
model_3 = nn.Sequential( nn.Linear(2,1), nn.ReLU() )

n_points = 100

X, x1,x2 = generateXGrid(n_points,50)

Z1 = model_1(X).detach().numpy()
Z2 = model_2(X).detach().numpy()
Z3 = model_3(X).detach().numpy()



import plotly as plo
import plotly.graph_objects as go
from plotly.subplots import make_subplots

# create figure 
fig = make_subplots(rows=1, cols=3,subplot_titles=('Model 1', 'Model 2', 'Model 3'), specs=[[{'type': 'scene'}] * 3])

fig.add_trace(go.Surface(z=Z1.reshape(n_points,n_points), x=x1, y=x2),row=1,col=1)
fig.add_trace(go.Surface(z=Z2.reshape(n_points,n_points), x=x1, y=x2),row=1,col=2)
fig.add_trace(go.Surface(z=Z3.reshape(n_points,n_points), x=x1, y=x2),row=1,col=3)


fig.show()

```
